```python
from godotbridge import GodotBridge

robot = GodotBridge(4243)
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import time
```

Avance en ligne droite du robot

```python
velocity = 5
robot.call("set_right_wheel_vel", velocity)
robot.call("set_left_wheel_vel", velocity)
time.sleep(2)
robot.call("set_right_wheel_vel", 0)
robot.call("set_left_wheel_vel", 0)
```

Enregistrement des donn√©es du robot

```python
velocity = 5
fps = robot.call("get_fps")
record_time = 5
robot.call("setup_record", record_time)
robot.call("set_right_wheel_vel", -velocity)
robot.call("set_left_wheel_vel", velocity)
time.sleep(3)
robot.call("set_right_wheel_vel", 0)
robot.call("set_left_wheel_vel", 0)
```

```python
right_vel = robot.call("get_right_wheel_vel")
left_vel = robot.call("get_left_wheel_vel")
```

```python
x_pos = robot.call("get_x_pos")
y_pos = robot.call("get_y_pos")
```

```python
plt.figure(1, clear = True)
plt.subplot(2,1,1)
t = np.linspace(0, record_time, record_time*fps)
plt.plot(t, right_vel)
plt.plot(t, left_vel)
plt.grid()
plt.subplot(2,1,2)
plt.plot(t, x_pos, label="x")
plt.plot(t, y_pos, label="y")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(2, clear=True, figsize=(3,2))
plt.plot(x_pos, y_pos)
plt.xlim(0, 3)
plt.ylim(0,2)
plt.grid()
plt.show()
```
